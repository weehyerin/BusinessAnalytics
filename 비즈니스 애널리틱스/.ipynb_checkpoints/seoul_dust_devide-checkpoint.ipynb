## 서울시 미세먼지 데이터로 미세먼지 있는 날 없는 날 구분

```python
df = pd.read_csv('20180401_20190331.csv', encoding='CP949')
df.head(10)
```

```python
df = df.dropna(axis=0)
df.head(10)
```

```python
date = df['날짜'].tolist()
date
```

```python
def replaceRight(original, old, new, count_right):
    repeat=0
    text = original
    
    count_find = original.count(old)
    if count_right > count_find : # 바꿀 횟수가 문자열에 포함된 old보다 많다면
        repeat = count_find # 문자열에 포함된 old의 모든 개수(count_find)만큼 교체한다
    else :
        repeat = count_right # 아니라면 입력받은 개수(count)만큼 교체한다

    for _ in range(repeat):
        find_index = text.rfind(old) # 오른쪽부터 index를 찾기위해 rfind 사용
        text = text[:find_index] + new + text[find_index+1:]
    
    return text
```

```python
n = len(date)
n
for i in range(0, len(date)):
    check = date[i]
    for j in range(10, 32):
        replace = str(j)
        prev = '0'+ replace
        check = check[::-1].replace(prev[::-1], replace[::-1], 1)[::-1]
        check = check.replace("218", "2018",1)
        check = check.replace("219", "2019",1)
    date[i] = check
```

```python
df["dust"] = pd.to_numeric(df["dust"])
df.head(10)
```

```python
df.loc[df.dust>35, 'dust'] = '나쁨'
df.head(10)
```

```python
df.loc[df.dust!='나쁨', 'dust'] = '안나쁨'
df.head(10)
```

```python
date = df['날짜'].tolist()
dust = df['dust'].tolist()
```
