```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

bad_circulation = pd.read_csv("bad_circulation.csv")
soso_circulation = pd.read_csv("soso_circulation.csv")

del bad_circulation['Unnamed: 0']
del soso_circulation['Unnamed: 0']
```

```python
bad_date = bad_circulation.date
#home living 부분만 column 이름에 따라서 바꿔주면 됨. 원하는 거
bad_hobby = bad_circulation.hobby
soso_date = soso_circulation.date
soso_hobby = soso_circulation.hobby
```

```python
bad_date = pd.DataFrame(bad_date)
bad_hobby = pd.DataFrame(bad_hobby)
soso_date = pd.DataFrame(soso_date)
soso_hobby = pd.DataFrame(soso_hobby)
```

```python
bad = pd.concat([bad_date, bad_hobby], axis = 1)
soso = pd.concat([soso_date, soso_hobby], axis = 1)

bad = bad.rename(columns={'hobby': 'bad_hobby'})
soso = soso.rename(columns={'hobby': 'soso_hobby'})
bad = bad.reset_index(drop=True)
soso = soso.reset_index(drop=True)
```

```python
soso = soso.dropna()
bad = bad.dropna()
```

```python
bad_date = bad.date.tolist()
bad_hobby = bad.bad_hobby.tolist()
soso_date = soso.date.tolist()
soso_hobby = soso.soso_hobby.tolist()

bad_column = bad_circulation.columns
result = []
for i in bad_hobby:
    
    text = str(i)
    text= text.replace("%","")
    result.append(text)
    
bad_hobby = pd.DataFrame(result)
bad_date = pd.DataFrame(bad_date)

bad = pd.concat([bad_date, bad_hobby], axis = 1)
bad.columns = ["date", "bad_hobby"]

result = []
for i in soso_hobby:
    
    text = str(i)
    text= text.replace("%","")
    result.append(text)
    
soso_hobby = pd.DataFrame(result)
soso_date = pd.DataFrame(soso_date)

soso = pd.concat([soso_date, soso_hobby], axis = 1)
soso.columns = ["date", "soso_hobby"] 

bad_ha = {}
result = []
date = bad.date
hobby = bad.bad_hobby
for i in range(0, len(date)):
    result = []
    for j in range(i, len(date)):
        if i == j:
            if date[i] == date[j]:
                if(date[i] in bad_ha):
                    if type(bad_ha[date[i]]) != list:
                        result.append(bad_ha[date[i]])
                    else:
                        result = bad_ha[date[i]]
                    result.append(hobby[i])
                    bad_ha[date[i]] = result
                    pass
                else:
                    bad_ha[date[i]] = hobby[i]
            else:
                i = j
                break
        else:
            i = j
            break
            
bad_date_key = list(bad_ha.keys())
bad_hobby_aver = {}
hobby_result_bad = []
hobby_date_bad = []

for i in bad_date_key:
    aver = 0
    for j in bad_ha[i]:
        aver = aver + float(j)
    aver = aver / len(bad_ha[i])
    hobby_result_bad.append(aver)
    hobby_date_bad.append(i)
    bad_hobby_aver[i] = aver
    
soso_ha = {}
result = []
date = soso.date
hobby = soso.soso_hobby
for i in range(0, len(date)):
    result = []
    for j in range(i, len(date)):
        if i == j:
            if date[i] == date[j]:
                if(date[i] in soso_ha):
                    if type(soso_ha[date[i]]) != list:
                        result.append(soso_ha[date[i]])
                    else:
                        result = soso_ha[date[i]]
                    result.append(hobby[i])
                    soso_ha[date[i]] = result
                    pass
                else:
                    soso_ha[date[i]] = hobby[i]
            else:
                i = j
                break
        else:
            i = j
            break

soso_date_key = list(soso_ha.keys())
soso_hobby_aver = {} 
hobby_result_soso = []
hobby_date_soso = []

for i in soso_date_key:
    aver = 0
    for j in soso_ha[i]:
        aver = aver + float(j)
    aver = aver / len(soso_ha[i])
    hobby_result_soso.append(aver)
    hobby_date_soso.append(i)
    soso_hobby_aver[i] = aver
    
hobby_date_soso_temp = hobby_date_soso
hobby_date_soso_temp = pd.DataFrame(hobby_date_soso_temp)
hobby_result_soso_temp = hobby_result_soso
hobby_result_soso_temp = pd.DataFrame(hobby_result_soso_temp)
soso_hobby = pd.concat([hobby_date_soso_temp, hobby_result_soso_temp], axis=1)
soso_hobby.columns = ['date', 'aver_soso']

hobby_date_bad_temp = hobby_date_bad
hobby_date_bad_temp = pd.DataFrame(hobby_date_bad_temp)
hobby_result_bad_temp = hobby_result_bad
hobby_result_bad_temp = pd.DataFrame(hobby_result_bad_temp)
bad_hobby = pd.concat([hobby_date_bad_temp, hobby_result_bad_temp], axis=1)
bad_hobby.columns = ['date', 'aver_bad']

hobby = pd.merge(soso_hobby, bad_hobby, how='outer')
hobby = hobby.sort_values(by = "date")
hobby = hobby.reset_index()

del hobby['index']

hobby['DateTime'] = hobby['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
del hobby['date']
hobby = hobby.set_index("DateTime")
```

```python
hobby.plot()
plt.ylabel("Data")
plt.show()
```
