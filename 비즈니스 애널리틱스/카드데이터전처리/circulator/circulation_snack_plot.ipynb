```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("snack_aver.csv")
```

```python
df.columns
```

```python
date = df['date']
```

```python
soso_aver = df['aver_soso']
```

```python
soso = pd.concat([date, soso_aver], axis=1)
```

```python
soso_date = soso['date']
aver_soso = soso['aver_soso']
```

```python
df['DateTime'] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
del df['date']
```

```python
df = df.set_index("DateTime")
```

```python
df.plot()
plt.ylabel("Data")
plt.rcParams["figure.figsize"] = [25,9]
plt.show()
```

```python
plt.plot(soso_date, aver_soso)
plt.xticks(rotation=90)
plt.show()
```
