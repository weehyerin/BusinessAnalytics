```python
import pandas as pd
```

```python
df = pd.read_csv('circulation.csv', encoding='CP949')
df.head(10)
```

```python
df.columns = ['date','gu', 'dong', 'sales', 'meal', 'snack', 'drink', 'home_living', 'health_beauty', 'hobby', 'social','infant']
```

```python
df.head(10)
```

```python
dust = []
with open('bad.txt', 'r') as f:
    for line in f:
        replace = line.replace('\n', '', 1)
        dust.append(replace)
        #df = df.loc[df.OPER_DT == line, 'OPER_DT'] = '나쁨'
```

```python
df["date"] = pd.to_numeric(df["date"])
df.head(10)
```

```python
df.date.dtype
```

```python
date = df['date'].tolist()
```

```python
dust = list(map(int, dust))

print(type(dust[1]))
```

```python
for i in range(0, len(dust)):
    comp = dust[i]
    for j in range(0, len(date)):
        if comp == date[j]:
            date[j] = '나쁨'
for j in range(0, len(date)):
    if date[j] != '나쁨':
        date[j] = '안나쁨'
```

```python
df['dust'] = date
df
```

```python
df.to_csv("circulation_dust.csv", mode='w')
```
