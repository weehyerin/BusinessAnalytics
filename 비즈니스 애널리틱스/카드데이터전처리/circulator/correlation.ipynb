```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

bad_circulation = pd.read_csv("bad_circulation.csv")
soso_circulation = pd.read_csv("soso_circulation.csv")

del bad_circulation['Unnamed: 0']
del soso_circulation['Unnamed: 0']
```

```python
bad_circulation['dust'] = 1
soso_circulation['dust'] = 0
```

```python
total = pd.concat([bad_circulation, soso_circulation])
```

```python
total.columns
```

```python
del total['date']
```

```python
total = total.reset_index()
```

```python
column = total.columns
ha = {}
for i in column:
    result = []
    test = total[i]
    for j in range(len(test)):
        text = str(test[j])
        text= text.replace("%","")
        result.append(text)
    ha[i] = result
```

```python
sales = ha['sales']
sales = [float(i) for i in sales]
sales = pd.DataFrame(sales)
meal = ha['meal']
meal = [float(i) for i in meal]
meal = pd.DataFrame(meal)
snack = ha['snack']
snack = [float(i) for i in snack]
snack = pd.DataFrame(snack)
drink = ha['drink']
drink = [float(i) for i in drink]
drink = pd.DataFrame(drink)
home_living = ha['home_living']
home_living = [float(i) for i in home_living]
home_living = pd.DataFrame(home_living)
health_beauty = ha['health_beauty']
health_beauty = [float(i) for i in health_beauty]
health_beauty = pd.DataFrame(health_beauty)
hobby = ha['hobby']
hobby = [float(i) for i in hobby]
hobby = pd.DataFrame(hobby)
social = ha['social']
social = [float(i) for i in social]
social = pd.DataFrame(social)
infant = ha['infant']
infant = [float(i) for i in infant]
infant = pd.DataFrame(infant)
```

```python
total = pd.concat([sales, meal, snack, drink, home_living, health_beauty, hobby, social, infant],axis=1)
```

```python
dust = ha['dust']
dust = [int(i) for i in dust]
dust = pd.DataFrame(dust)
```

```python
total = pd.concat([total, dust], axis=1)
```

```python
total.columns = ['sales', 'meal', 'snack', 'drink', 'home_living', 'health_beauty', 'hobby', 'social', 'infant', 'dust']
```

```python
col = total.columns
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

fitted = std_scaler.fit(total)
print(fitted.mean_)
## 출력 결과
## array([4.46000000e+02, 3.83838384e-01, 2.30864198e+00, 2.93615825e+01,
##        5.23007856e-01, 3.81593715e-01, 3.22042080e+01])
total = std_scaler.transform(total)
total = pd.DataFrame(total, columns=col)
print(total.head())
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns   

plt.figure(figsize=(15,15))
sns.heatmap(data = total.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Blues')
```
