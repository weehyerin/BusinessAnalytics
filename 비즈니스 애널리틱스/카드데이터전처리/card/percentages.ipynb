```python
import pandas as pd

df = pd.read_csv("card_result.csv")
```

```python
df.columns
```

```python
value = df['0']
```

```python
total = value[0]
```

```python
percent = []
for i in range(len(value)):
    percent.append(value[i]/total*100)
```

```python
percent
```

```python
df['percent'] = percent
```

```python
df.to_csv('card_percent.csv')
```
